```python
from abc import ABC, abstractmethod

class Pagamento(ABC):
  @abstractmethod
  def validar(self):
    pass

  @abstractmethod
  def processar(self):
    pass

  def registrar(self):
    print("ENTREGAR")

  def executar(self):
    self.validar()
    self.processar()
    self.registrar()

class PagamentoPix(Pagamento):
  def validar(self):
    print("validando chave Pix")

  def processar(self):
    print("Transferindo")
```
